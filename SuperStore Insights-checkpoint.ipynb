```python
data.rename(columns = {'Sales' : 'Unit Price'} , inplace = True)
data['TotalSales'] = (data['Unit Price'] * data['Quantity']) + data['Discount']
```

```python
data['DisAmount'] = (data['Unit Price'] * data['Quantity']) * data['Discount']
data['DisAmount']
```

```python
data['TotalSales'] = (data['Unit Price'] * data['Quantity']) + data['DisAmount']
```

```python
data.insert(20 , 'Discount Amount' , data['DisAmount'] )
```

```python
pd.set_option('display.max.columns' , None)
```

```python
data.drop( 'DisAmount' ,axis = 1 , inplace = True)
```

```python
data['profits'] = data['TotalSales'] - data['Discount Amount']
data.head(2)
```

```python
data.rename(columns = {'TotalSales':'Total Sales' , 'profits' : 'Profit' } , inplace = True)
```

```python
data.drop( 'Profit' ,axis = 1 , inplace = True)
```

```python
gby_ship_mode = data.groupby('Ship Mode')['Total Sales'].sum().sort_values()
gby_ship_mode
```

```python
data.head(2)
```

```python
top_cus = data.groupby('Customer Name')['Total Sales'].sum().sort_values(ascending = False).head(10)
top_cus
```

```python
gby_country = data.groupby('Country')['Total Sales'].sum()
```

```python
gby_reg = data.groupby('Region')['Total Sales'].sum().sort_values()
gby_reg
```

```python
plt.figure(figsize = (20,5) , facecolor = '#1b2021')
plt.bar(gby_reg.index , gby_reg.values , width = 0.3 , color = '#51513d')
plt.plot(gby_reg , color = '#1b2021' , linestyle = '--', marker = 'o' , markersize  = 8)
plt.xticks(c = 'w' ,weight = 'bold' , rotation = 14)
plt.yticks(c = 'w' ,weight = 'bold' , fontsize = 12)
plt.title('Total Sales For Each Region' , pad = 20 , c = 'w' , fontsize = 25 , weight = 'bold')
plt.xlabel('Region' , weight = 'bold' , color  = 'w' , labelpad = 20 , fontsize = 15)
plt.grid(axis = 'x' ,linestyle = '--', color='#51513d')
offset = 0.1 * max(gby_reg.values)
for ind  , v  in zip(gby_reg.index , gby_reg.values) : 
    plt.text(ind , v - offset  , f'{v:.0f}' ,  ha =  'center' , va = 'center' , color = 'w' , fontsize = 11 ,weight = 'bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (19,5) , dpi = 100 , facecolor = '#606c38')
plt.bar(gby_ship_mode.index , gby_ship_mode.values , color = '#283618' , width = 0.5)
plt.plot(gby_ship_mode , color = '#606c38' , marker = 'o' , markersize = 10)
plt.title('Total Sales By Segment' , fontsize = 25, weight = 'bold' , color = 'w' , pad = 20)
plt.xticks(weight = 'bold' , color = 'w')
plt.yticks(color  = 'w')
plt.xlabel('Segment' , color = 'w' , weight = 'bold')
plt.grid(axis = 'x' , linestyle = '--',color = '#283618')
offset = 0.09 * max(gby_ship_mode.values)
for ind , v in zip(gby_ship_mode.index , gby_ship_mode.values) :
    plt.text(ind , v - offset   , f'{v:.0f}' , ha = 'center' , va = 'bottom' ,color = 'w'  , weight = 'bold')
plt.show()
```

```python
plt.figure(figsize = (20,7) , dpi = 100 , facecolor = '#621708')
plt.barh(top_cus.index , top_cus.values , color = '#220901' )
plt.yticks(color = 'w' , weight = 'bold' ,rotation = 10)
plt.xticks(color = 'w' , weight = 'bold')
plt.grid(axis = 'y' , color = 'w' , linestyle = '--')
plt.title('Top selling customers' , color = 'w' , pad = 20 , fontsize = 25 , weight = 'bold'
         , bbox = dict(facecolor = '#220901'))
offset = 0.01 * max(top_cus.values)
for i , v in zip(top_cus.index , top_cus.values):
    plt.text(v+ offset , i  , f'{v:.0f}' , ha  = 'left' , va = 'top' , color = '#220901' , weight = 'bold')
plt.show()
```

```python
od_data = data.groupby(data['Order Date'].dt.year)['Total Sales'].sum()
sh_data = data.groupby(data['Ship Date'].dt.year)['Total Sales'].sum()
```

```python
plt.figure(figsize = (20 , 5) , dpi = 100 , facecolor = '#004346'  )
plt.plot(od_data , linewidth = 8 , marker = 'o' , markersize = 15  , color = '#004346' , label='Order Data')
plt.plot(sh_data , linestyle = '--' , color  = '#75dddd' , marker = '*' , markersize = 10 , label='Ship Data')
plt.fill_between(od_data.index , od_data.values , color = '#004346' , alpha = 0.5)
plt.xticks(sh_data.index , c = 'w' , fontsize = 15)
plt.grid(axis = 'x' , linestyle = '--')
plt.legend()
plt.title('Total Sales By Order Date And Ship Date' , color = 'w' , weight = 'bold' , fontsize = 25,  pad = 25
         , bbox = dict(facecolor = 'black'))
offset = 0.07 * max(sh_data.values)
for ind , v in zip(sh_data.index , sh_data.values) :
    plt.text(ind , v + offset , f'{v:.0f}' , ha = 'center' , va = 'center' , weight = 'bold' , color = '#75dddd' , 
             bbox=dict(facecolor='black', edgecolor='black', boxstyle='round,pad=0.3'))
for ind , v in zip(od_data.index , od_data.values) :
    plt.text(ind , v - offset , f'{v:.0f}' , ha = 'center' , va = 'center' , weight = 'bold' , color = 'black')
plt.show()
```
